## Joining data 1
In deze opdracht ga je aan de slag met het samenvoegen van data uit meerdere bronnen. De csv-bestanden zijn afkomstig van een Kaggle competitie waarin de deelnemers zo goed mogelijk de toekomstige verkopen van Walmart proberen te voorspellen: https://www.kaggle.com/c/walmart-recruiting-store-sales-forecasting

- Voer onderstaande cel uit om 3 dataframes aan te maken vanuit 3 csv-bestanden. 

```python
import pandas as pd
stores = pd.read_csv('stores.csv')
sales = pd.read_csv('sales.csv')
features = pd.read_csv('features.csv')
```

### 1 Inspecteren
- Inspecteer de 3 dataframes:
    - Laat zien welke kolommen er zijn en wat het datatype is van iedere kolom.
    - Laat zien hoeveel rijen ieder dataframe heeft.
    - Laat de eerste 5 rijen zien van ieder dataframe.
    - Noteer hoeveel rijen je verwacht als je sales en stores samenvoegd met een 'inner join'.
    - Noteer hoeveel rijen je verwacht als je sales, stores en features samenvoegd met een 'inner join'.

```python
print('Stores:',stores.columns)
print('Sales',sales.columns)
print('Features',features.columns)

print('Stores:',len(stores.index))
print('Sales',len(sales.index))
print('Features',len(features.index))

display('Stores:',stores.head())
display('Sales',sales.head())
display('Features',features.head(5))

# 421570
# 421570
```

### 2 Samenvoegen van sales and stores
- Gebruik een 'inner join' om sales en stores samen te voegen in een nieuw dataframe sales_stores.
    - Hint: bepaal eerst op welke kolom(men) je de dataframes kunt samenvoegen.
- Laat de eerste 5 rijen zien van sales_stores.
- Laat zien hoeveel rijen sales_stores bevat. Komt dit overeen met wat je verwachtte?

```python
sales_stores = sales.merge(stores, how= "inner", left_on='Store', right_on='Store')
display (sales_stores.head())
display(len(sales_stores.columns))
```

### 3 Analyseren samengevoegd dataframe
- Toon de gemiddelde verkopen (Weekly_sales) en de gemiddelde grootte (Size) per winkeltype (Type). Hint: maak gebruik van groupby().

```python
import numpy as np

display (sales_stores.groupby('Type')[['Weekly_Sales', 'Size']].agg(np.mean))
```

## 4 Samenvoegen van sales, stores and features 
- Gebruik een 'inner join' om sales, features en stores samen te voegen in een nieuw dataframe sales_stores_features.
    - Hint: bepaal eerst op welke kolom(men) je de dataframes kunt samenvoegen.
- Gebruik 'suffixes' om te laten zien uit welk dataframe een kolom afkomstig is.

```python


sales_stores_features = sales_stores.merge(features, on=['Store', 'Date', 'IsHoliday'],  how='left', suffixes=('_Ss','_Fe'))

def renameColumns(database, suffix='Suff', drop_columns = ""):
    columns = database.columns.drop(drop_columns)
    db_return = dict()
    for col in  columns:
        db_return[col] = col + suffix
    return db_return



sales_stores_features.rename(columns = renameColumns(sales, '_Sales', ['Store','Date', 'IsHoliday']), inplace=True)

sales_stores_features.rename(columns = renameColumns(stores, '_Stores', ['Store']), inplace=True)

sales_stores_features.rename(columns = renameColumns(features, '_Features', ['Store','Date']), inplace=True)

display(sales_stores_features)


```

### 5 Samenvoegen zonder suffixes.
- Er is een mogelijkheid om de 3 dataframes samen te voegen zonder dat er 'suffixes' nodig zijn. Voeg de dataframes samen op een wijze dat er geen suffixes nodig zijn.
    - Hint: je moet hierbij samenvoegen op 3 kolommen.

```python
sales_stores = sales.merge(stores, how= "left", left_on='Store', right_on='Store')
sales_stores_features = pd.merge(sales_stores,features, left_on=['Store', 'Date', 'IsHoliday'], right_on=['Store', 'Date', 'IsHoliday'], how='left')
display(sales_stores_features)
```

### 6 Analyseren met een draaitabel
- Maak een draaitabel waarin je de gemiddelde verkopen (Weekly_Sales) laat zien per winkeltype (Type) en IsHoliday. Zorg ervoor dat ook het totale gemiddelde wordt getoond.

```python
import numpy as np
display(sales_stores_features.groupby(['Type','IsHoliday'])['Weekly_Sales'].agg(np.mean))
```

```python

sales_stores_piv = sales_stores_features.pivot_table(values = 'Weekly_Sales', index = ['Type','IsHoliday'],aggfunc= {'Weekly_Sales':np.mean}, margins = True) 
    
display(sales_stores_piv)
```
