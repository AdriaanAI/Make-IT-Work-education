Import libraries

```python
import pandas as pd
import numpy as np
```

Set variables and import file in a DataFrame 'data'

Show first 5 rows of 'data'

```python
filename = 'onderwijs_uitgave.xlsx'
data = pd.read_excel(filename, na_values='.')

display(data.head())

# data.loc[data['Waarde'].str.isdecimal()==False,'Waarde'] = 0

# data = data.convert_dtypes(convert_string=False, convert_integer=True)


display(data.info())
```

Bepaal door het selecteren van data uit een dataset in Python de uitgaven aan primair basisonderwijs in 1989.

```python
data_1989 = data[data['Perioden']==1989]

uitgave_1989 = data_1989.loc[data_1989['Type onderwijs'].str.lower().str.contains('primair onderwijs'),'Waarde'].sum()

print ('Uitgave Primaiere basisonderwijs =', uitgave_1989 )
```


Bepaal door middel van Python code de toename van de overheidsguitgaven aan onderwijs voor het onderwijstype wetenschappelijk onderwijs in de periode 2000-2010

```python
data_2000_2010 = data[ (data['Perioden'] >=2000) & 
                      ( data['Perioden'] <=2010)]

uitgave_2000_2010 = data_2000_2010[data_2000_2010['Type onderwijs'].str.lower().str.contains('wetenschappelijk onderwijs') &
                       data_2000_2010['Type uitgave'].str.lower().str.contains('overheidsuitgaven aan onderwijs')]

toename_uitgave_2000_2010 = uitgave_2000_2010.loc[uitgave_2000_2010['Perioden']== 2010, 'Waarde'].values[0] -  uitgave_2000_2010.loc[uitgave_2000_2010['Perioden'] == 2000, 'Waarde'].values[0]



print ('Toename uitgaven van 2000 tot 2010 is:', toename_uitgave_2000_2010)
```

Selecteer alleen de overheidsuitgaven aan onderwijs in de kolom Type uitgaven sla de data op in een tabel met de naam “overheidsuitgaven"

```python

overheidsuitgaven = data[data['Type uitgave'] == 'Overheidsuitgaven aan onderwijs']

display (overheidsuitgaven.head())
```

Gebruik de nieuwe tabel en maak een draaitabel met in de rijen de Perioden in de kolommen het type onderwijs. Welke manier van aggregeren wordt automatisch gekozen?

```python
# draai_overheidsuitgaven = 
draai_overheidsuitgaven = pd.pivot_table(overheidsuitgaven, index='Perioden', columns='Type onderwijs', values= 'Waarde')

display(draai_overheidsuitgaven)

# mean
```

Maak nu een draaitabel met perioden in de rijen, geen kolommen en als aggregeerfunctie ‘max’. Dit geeft per jaar de maximale uitgave per type onderwijs.

```python
draai_perioden_maxn = pd.pivot_table(overheidsuitgaven, index='Perioden', values= 'Waarde', aggfunc=('max'))

display(draai_overheidsuitgaven)
```
