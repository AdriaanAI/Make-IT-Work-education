### 1.3 Boolean operators: and, or, not
1) Maak print statements waarmee je checkt of:
- stamford_staff groter is dan 6 EN kleiner is dan 8
- stamford_staff groter is dan 6 OF kleiner is dan 8

2) Pas de not() functie toe om dezelfde checks nogmaals te doen.

```python
stamford_staff = 8
skipton_staff = 2
#start coding here
print ('Stamford staff over 6 and smaller then 8 is', (stamford_staff > 6) & (stamford_staff < 8))
print ('Stamford staff smaller then 6 or over 8 is', (stamford_staff > 6) | (stamford_staff < 8))

print('\nResult with not():')
print ('Stamford staff over 6 and smaller then 8 is', not((stamford_staff > 6) & (stamford_staff < 8)))
print ('Stamford staff smaller then 6 or over 8 is', not((stamford_staff > 6) | (stamford_staff < 8)))
```

### 2.1 if, else, elif

**stamford_staff**
- Print 'big store' als stamford_staff groter is dan 8. 
- Print 'medium store' als stamford_staff groter is dan 6 en kleiner of gelijk aan 8.
- Print 'small store' in alle andere gevallen.

**stamford_loc**
- Print stamford_loc als de locatie gelijk is aan 'Retail Park'. 
- Als dit niet het geval is print dan stamford_loc als de locatie gelijk is aan 'Shopping Centre'. 
- Print de eerste letter van de woorden in stamford_loc met hoofdletters als de eerdere checks niet waar zijn.

```python
stamford_staff = 8
stamford_loc = "shopping centre"

# Start coding here: stamford_staff
if stamford_staff > 8:
    print ('big store')
elif stamford_staff > 6:
    print ('medium store')
else:
    print ('small store')
    
# Start coding here: stamford_loc
if stamford_loc == 'Retail Park':
    print ('location',stamford_loc, 'is equal to Retail Park')
elif stamford_loc == 'Shopping Centre':
    print ('location',stamford_loc, 'is equal to Shopping Centre')
else:
    print (stamford_loc.capitalize())
```
