Import libraries and data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('Dataset1 week 10.xlsx')
```

Show info on df DataFrame

```python
display(df.head())
display(df.info())
display(df.shape)
```

```python
#Check Data
df['one']=1


df.drop(df[(df['Marital_Status']=='Absurd')|
           (df['Marital_Status']=='Alone')].index, inplace=True)
display(df.describe())
display(df[df['Income']<6000])
display(df[df['Income']>100000])

df['Marital_Status'].replace('Alone','Single', inplace=True)

#Change Dtype of Education & Marital_Status to Category

df['Education'] = df['Education'].astype('category')
df['Marital_Status'] = df['Marital_Status'].astype('category')
df['Kidhome'] = df['Kidhome'].astype('category')
df['Teenhome'] = df['Teenhome'].astype('category')

```

```python
# The Responded to deal compared to Income

sns.scatterplot(data=df, hue='Responded to deal', y='Income', x='Year_Birth')
```

```python
# The Responded to deal compared to Income

sns.barplot(data=df, y= 'Income', x='Education', errorbar=None, palette='Blues')
```

```python
sns.barplot(data=df, y='Responded to deal', x='Marital_Status', estimator='mean', errorbar=None, palette='Blues')
plt.title('Mean responded to deal')
plt.ylabel('responded to deal')
plt.show()
```

```python
sns.barplot(data=df, hue='Responded to deal', y= 'one', x='Education', estimator=sum, palette='Blues')
plt.title('Number of Responded to deal by Martial_status')
plt.ylabel('# People')
plt.show()
```

```python
stats = df.pivot_table(index=['Responded to deal', 'Education', 'Marital_Status'],
                       values=['Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds'],
                       aggfunc=np.mean)

display(stats)
sns.heatmap(stats)
```
