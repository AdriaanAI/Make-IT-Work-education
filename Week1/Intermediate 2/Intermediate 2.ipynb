# Dictionaries & Pandas
- Voer onderstaande cel uit om de st_to dictionary aan te maken.

```python
#Creating dictionary from columns in dataframe
import pandas as pd
storedata = pd.read_csv('storedata_clean.csv',sep=";")

store_id = storedata['Store ID'].values.tolist()
turnover = storedata['Profit'].values.tolist()
floor = storedata['Floor Space'].values.tolist()

st_to = dict(zip(store_id, turnover))
```

### 1.1 Printen dictionary
St_to bevat de winkelnummers (store id) en de omzet (turnover) van iedere winkel.
- Print st_to.

```python
print(st_to)
```

## 2 Single dictionaries
Opdrachten:

- Print de omzet van winkel 13132
- Voeg winkel 99999 met een omzet van 3100000 toe aan de dictionary. 
- Print de omzet van 99999.
- Vervang de omzet van 99999 met 3200000 en print de nieuwe omzet.
- Verwijder winkel 99999 en controleer dit door te testen of je de omzet nog kunt printen.

```python
#print turnover of store 13132
print (st_to.get(13132))

#Add store 99999 with turnover of 3100000. Print the turnover
st_to[99999] = 3100000

#print turnover of store 13132
print (st_to.get(99999))

#Update turnover of the new store to 3200000 and print the turnover
st_to[99999] = 3200000
print (st_to.get(99999))

#delete the new store and check deletion by attempting to print it again
del(st_to[99999])
print (st_to.get(99999))
```

## 3 Nested dictionaries
- Voer onderstaande cel uit om de dictionary st_to_fl aan te maken.

```python
#Creating a nested dictionary
st_to_fl = dict(map(lambda x, y, z : (x, {"turnover":y , "floor":z}), store_id, turnover, floor))
```

- Print st_to_fl en bestudeer de structuur van deze 'nested' dictionary.

```python
#Get an understanding of the structure of the dictionary
print(st_to_fl)
```

- Print de oppervlakte van 13132.
- Voeg de nieuwe winkel 99999 toe aan de dictionary met een omzet van 3100000 en een oppervlakte van 15000.
- Print de oppervlakte en de omzet van de nieuwe winkel.
- Verander de oppervlakte van 99999 in 3200000.
- Print de omzet van de nieuwe winkel.

```python
#print floorspace of 13132
print(st_to_fl[13132]['floor'])

#add new store 99999 to the dictionary with turnover of 3100000 and floor space of 15000
st_to_fl[99999] = {'turnover': 3100000,'floor': 15000}

#Change turnover into 3200000
st_to_fl[99999]['turnover'] = 320000
print(st_to_fl)
```
